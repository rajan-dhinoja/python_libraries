### How Numpy occupies less space than Python

```python
import sys
```

```python
py_arr = [1,2,3,4,5]
```

```python
np_arr = np.array(py_arr)
```

```python
print(sys.getsizeof(1))
print(len(py_arr))
sys.getsizeof(1) * len(py_arr)
```

```python
print(np_arr.itemsize)
print(np_arr.size)
np_arr.itemsize * np_arr.size
```
