```python
x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']
plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)
# plt.bar(x,y,color=c, width=0.4)
# plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5)
# plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5, linestyle=":")
plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5, linestyle=":", alpha=0.5)
plt.show()
```

```python
plt.bar(x,y,color=c, width=0.4, align="edge", linewidth=5, label='popularity1')
plt.bar(x,z, width=0.4, align="edge", linewidth=5, label='popularity2')
plt.legend()
plt.show()
```

```python

x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']

p = np.arange(len(x))
width=0.2
p1= [j+width for j in p]

plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)

plt.bar(p,y,width, color=c, align="edge", linewidth=5, label='popularity1')
plt.bar(p1,z,width,  align="edge", linewidth=5, label='popularity2')

# plt.xticks(p1, x)
plt.xticks(p+width, x, rotation=30)
plt.legend(loc='upper center')
plt.show()
```

```python

x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']

p = np.arange(len(x))
width=0.8
p1= [j+width for j in p]

plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)

plt.bar(p,y,width, color=c, align="edge", linewidth=5, label='popularity1')
plt.barh(p1,z,width,  align="edge", linewidth=5, label='popularity2')

# plt.xticks(p1, x)
plt.xticks(p+width, x, rotation=30)
plt.legend(loc='upper center')
plt.show()
```
